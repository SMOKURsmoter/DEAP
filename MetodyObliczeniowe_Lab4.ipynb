```python
import numpy as np
import matplotlib.pyplot as plt

# Tworzenie obiektu wielomianu Czebyszewa
cheb_poly = np.polynomial.Chebyshev([1, 2, 3])

# Generowanie punktów do wykresu
x = np.linspace(-1, 1, 100)
y = cheb_poly(x)

# Rysowanie wykresu
plt.plot(x, y)
plt.title('Wielomian Czebyszewa')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Chebyshev

# Funkcja do aproksymacji
def funkcja_aproksymowana(x):
    return np.sin(x)

# Punkty do aproksymacji
x_punkty = np.linspace(0, 2 * np.pi, 10)

# Wartości funkcji do aproksymacji
y_punkty = funkcja_aproksymowana(x_punkty)

# Stopień wielomianu Czebyszewa
stopien = 5

# Aproksymacja wielomianem Czebyszewa
cheb_approx = Chebyshev.fit(x_punkty, y_punkty, deg=stopien)

# Punkty dla wykresu aproksymacji
x_aproksymacja = np.linspace(0, 2 * np.pi, 100)
y_aproksymacja = cheb_approx(x_aproksymacja)

# Rysowanie wykresu
plt.plot(x_punkty, y_punkty, 'ro', label='Punkty aproksymacji')
plt.plot(x_aproksymacja, y_aproksymacja, label=f'Aproksymacja Czebyszewa (stopień {stopien})')
plt.title('Aproksymacja funkcji sinus za pomocą wielomianu Czebyszewa')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definicja funkcji
def funkcja(x):
    return np.exp(x) * np.cos(0.5 * x**2)

# Generowanie danych
x = np.linspace(0, 3, 1000)
y = funkcja(x)

# Rysowanie wykresu
plt.plot(x, y, label='y = exp(x) * cos(0.5 * x^2)')
plt.title('Wykres funkcji y = exp(x) * cos(0.5 * x^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def funkcja(x):
    return np.exp(x) * np.cos(0.5 * x**2)

def cheb_approx(x, degree):
    approx_values = np.polynomial.chebyshev.Chebyshev.fit(x, funkcja(x), degree)(x)
    return approx_values

# Generowanie danych
x = np.linspace(0, 3, 1000)
y = funkcja(x)

# Wykres funkcji oryginalnej
plt.plot(x, y, label='y = exp(x) * cos(0.5 * x^2)')

# Wykres aproksymacji Czebyszewa stopnia 10
r10 = cheb_approx(x, 2)
plt.plot(x, r10, label='Aproksymacja Czebyszewa stopień 2', linestyle='dashed')

# Wykres aproksymacji Czebyszewa stopnia 10
r10 = cheb_approx(x, 4)
plt.plot(x, r10, label='Aproksymacja Czebyszewa stopień 4', linestyle='dashed')

# Wykres aproksymacji Czebyszewa stopnia 10
r10 = cheb_approx(x, 6)
plt.plot(x, r10, label='Aproksymacja Czebyszewa stopień 6', linestyle='dashed')


# Ustawienia wykresu
plt.title('Wykres funkcji oraz aproksymacji Czebyszewa')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def funkcja(x):
    return 1 / (12 * x**2 + 1)

def cheb_approx(x, degree):
    approx_values = np.polynomial.chebyshev.Chebyshev.fit(x, funkcja(x), degree)(x)
    return approx_values

# Generowanie danych
x = np.linspace(-2, 2, 1000)
y = funkcja(x)

# Wykres funkcji oryginalnej
plt.plot(x, y, label=r'$y = \frac{1}{12x^2 + 1}$')

# Wykres aproksymacji Czebyszewa stopnia 10
r10 = cheb_approx(x, 10)
plt.plot(x, r10, label='Aproksymacja Czebyszewa stopień 10', linestyle='dashed')

# Wykres aproksymacji Czebyszewa stopnia 15
r10 = cheb_approx(x, 15)
plt.plot(x, r10, label='Aproksymacja Czebyszewa stopień 15', linestyle='dashed')

# Wykres aproksymacji Czebyszewa stopnia 20
r10 = cheb_approx(x, 20)
plt.plot(x, r10, label='Aproksymacja Czebyszewa stopień 20', linestyle='dashed')





# Ustawienia wykresu
plt.title('Wykres funkcji oraz aproksymacji Czebyszewa')
plt.xlabel('x')
plt.ylabel('y')

# Przesunięcie legendy w dół
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=2)

plt.grid(True)
plt.show()
```
